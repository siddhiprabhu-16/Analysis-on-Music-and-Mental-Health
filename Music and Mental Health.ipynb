```python
import numpy as np

df['Age'].value_counts().loc[lambda x : x>50]
```

```python
import matplotlib.pyplot as plt

s_colors = ['lightgreen', 'lightcoral', 'steelblue', 'palevioletred', 'gold', 'darkturquoise']

services = df['Music Platform'].value_counts()
services.plot(kind='pie', colors = s_colors)

plt.title('Streaming services by popularity')
plt.ylabel("")
```

```python
import seaborn as sns

s_colors2 = ['lightgreen', 'darkturquoise', 'lightcoral', 'steelblue', 'palevioletred', 'gold']

df.replace(['Streaming services (e.g., Spotify, Apple Music)', ' YouTube or other video platforms .', 'Personal music collection (e.g., MP3s)'],
                       ['Other', 'None', 'YouTube'], inplace=True)

bplot = sns.boxplot(data=df, x="Music Platform", y = "Age",
            showfliers = False,
            palette = s_colors2)

plt.title('Streaming services by Age')
```

```python
fig = plt.figure(figsize=(6,2))

plt.suptitle("Musical background")

ax = fig.add_subplot(121)

inst = df['Creating Music'].value_counts()
inst.plot(kind='pie', colors = ["indianred", "darkblue"], labeldistance = 1.2)
```

```python
plt.figure(figsize=(5,4))
plt.title('Effects of Music on Mental Health')

effects = df['Improvement'].value_counts()
effects.plot(kind='pie', colors = ["indianred", "gold", "darkblue"], ylabel= '');
```

```python
plt.figure(figsize=(5,4))
plt.title('Engage in creating music')

effects = df['Creating Music'].value_counts()
effects.plot(kind='pie', colors = ["orange", "brown", "blue"], ylabel= '');
```

```python
plt.figure(figsize=(5,4))
plt.title('Used as a Coping Mechanism for Mental Health')

effects = df['Coping Mechanism'].value_counts()
effects.plot(kind='pie', colors = ["indianred", "gold", "darkblue"], ylabel= '');
```
